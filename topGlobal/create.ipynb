```python
import pandas as pd
import glob

path = "/Users/emiliahernandez/Downloads"

files = sorted(glob.glob(f"{path}/regional-global-weekly-2025*.csv"))

print("Archivos encontrados:", len(files))

if len(files) == 0:
    raise ValueError("❌ No se encontraron CSV en el path indicado")

dfs = []

for f in files:
    print("Leyendo:", f)
    df = pd.read_csv(f)
    df["source_file"] = f.split("/")[-1]  # solo nombre
    dfs.append(df)

global_df = pd.concat(dfs, ignore_index=True)

output = f"/Users/emiliahernandez/Desktop/desarrollo/topGlobal/global2025.csv"
global_df.to_csv(output, index=False)

print("✅ Archivo creado:", output)
print("Filas totales:", len(global_df))
```
